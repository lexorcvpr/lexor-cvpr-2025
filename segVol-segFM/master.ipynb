```python
import numpy as np
import os

img_dir = '/home/sebastian/codes/data/CVPR-2025-CHALLENGE/3D_val_npz/3D_val_npz'
gt_dir = '/home/sebastian/codes/data/CVPR-2025-CHALLENGE/3D_val_gt/3D_val_gt/3D_val_gt_interactive'
output_dir = '/home/sebastian/codes/data/CVPR-2025-CHALLENGE/merged'

os.makedirs(output_dir, exist_ok=True)

missing, real_vals = 0, 0

for fname in os.listdir(img_dir):
    img_path = os.path.join(img_dir, fname)
    gt_path = os.path.join(gt_dir, fname)
    out_path = os.path.join(output_dir, fname)

    if not os.path.exists(gt_path):
        print(f"Missing ground truth for {fname}, skipping.")
        continue

    data_img = np.load(img_path, allow_pickle=True)
    data_gt = np.load(gt_path, allow_pickle=True)

    if 'boxes' not in data_img:
        print(f"⚠️  {fname} has no 'boxes'. Inserting empty list.")
        boxes = []
        missing += 1
    else:
        boxes = data_img['boxes']
        real_vals += 1
    np.savez(
        out_path,
        imgs=data_img['imgs'],
        boxes=boxes,
        spacing=data_img['spacing'],
        text_prompts=data_img.get('text_prompts', []),
        gts=data_gt['gts']
    )

    # np.savez_compressed( # --> cambiar porque commpressed no guarda archivos pesados 
    #     out_path,
    #     imgs=data_img['imgs'],
    #     boxes=boxes,
    #     spacing=data_img['spacing'],
    #     text_prompts=data_img.get('text_prompts', []),
    #     gts=data_gt['gts']
    # )

print(f"\n✅ Done. Total files processed: {real_vals + missing}")
print(f"   Files with boxes: {real_vals}")
print(f"   Files without boxes (fallback inserted): {missing}")
```

```python
import os
import json

# Ppath to the folder containing your .npz validation files
npz_folder = "/home/sebastian/codes/data/CVPR-2025-CHALLENGE/merged"

# Path where val_samples.json will be saved
output_json_path = "./val_samples.json"

# Collect all .npz files in the folder
val_samples = [
    os.path.join(npz_folder, fname)
    for fname in sorted(os.listdir(npz_folder))
    if fname.endswith(".npz")
]

# Save the list to a JSON file
with open(output_json_path, "w") as f:
    json.dump(val_samples, f, indent=2)

# Show basic info
print(f"val_samples.json has been saved at:\n{output_json_path}")
print(f"It contains {len(val_samples)} .npz files")

# Preview the first few entries
print("\n Preview:")
for entry in val_samples[:3]:
    print(f" - {entry}")
```
